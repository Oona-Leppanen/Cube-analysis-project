```python
# Initializing dataframes.

card_data = pd.read_csv('card_data.csv', delimiter = ';')
```

```python
# Takes user input.

print('READ!\n- Use the first letter of the color and use upper letter.')
print('- If you want to write multiple colors or types in the search filed use comma to separate values. E.g. write: R, G')
print('- If you dont want to use color or type in the search dont write anything.')
print('- Mana value is written using numbers not letters.')
print('- If you dont want to use mana value in the search write: 100.')
print(' - Use upper letters at the beginning of the type.')

color = list(map(str, input('Enter color: ').split(', ')))
mana = int(input('Enter mana value: '))
c_type = list(map(str, input('Enter type: ').split(', ')))

print(type(mana))
print(color)
```

```python
# Makes lists containing card names matching to input values.
# First creates some lists. Then adds card names to a list if card name matches given color/mana value/type.
# Then the lists are compared depending on are color, mana value and type given or is one of them not given.

color_list, mana_list, type_list = ([] for i in range(3))
color_mana, color_type, mana_type, color_mana_type = ([] for i in range(4))

for i in range(len(card_data)):
    if color != ['']:
        
        if color == ['Colorless']:
            for j in range(len(color)):
                if color[j] in card_data.iloc[i]['Color']:
                    if j == len(color)-1:
                        color_list.append(card_data.iloc[i]['Card_name'])
                else:
                    break
                    
        elif len(color) == len(card_data.iloc[i]['Color']):
            for j in range(len(color)):
                if color[j] in card_data.iloc[i]['Color']:
                    if j == len(color)-1:
                        color_list.append(card_data.iloc[i]['Card_name'])
                else:
                    break
                        
    if  mana != 100:
        if card_data.iloc[i]['Mana_value'] == mana:
            mana_list.append(card_data.iloc[i]['Card_name'])
            
    if c_type != ['']:
        if len(c_type) == len(card_data.iloc[i]['Type'].split(' ')):
            
            for x in range(len(c_type)):
                if c_type[x] in card_data.iloc[i]['Type']:
                    if x == len(c_type)-1:
                        type_list.append(card_data.iloc[i]['Card_name'])
                else:
                    break
            
# All inputs given.
if (color != ['']) & (mana != 100) & (c_type != ['']):
    for i in range(len(color_list)):
        for j in range(len(mana_list)):
            if color_list[i] == mana_list[j]:
                color_mana.append(color_list[i])
                
    for i in range(len(color_mana)):
        for j in range(len(type_list)):
            if color_mana[i] == type_list[j]:
                color_mana_type.append(color_mana[i])
                
    color_mana = [] # Resets the list.
    
# 2 inputs given.
elif color == ['']:
    for i in range(len(mana_list)):
        for j in range(len(type_list)):
            if mana_list[i] == type_list[j]:
                mana_type.append(mana_list[i])
                
elif mana == 100:
    for i in range(len(color_list)):
        for j in range(len(type_list)):
            if color_list[i] == type_list[j]:
                color_type.append(color_list[i])
                
elif c_type == ['']:
    for i in range(len(color_list)):
        for j in range(len(mana_list)):
            if color_list[i] == mana_list[j]:
                color_mana.append(color_list[i])
                
if (c_type != ['']) & (type_list == []):
    print('No cards with given features.')
```

```python
# Calculates winrate for searched cards.
# Checks which list contains at least one card name. If there's used just one value but the value contains at least 2 elements
# then that list will be used to calculate winrate. Mana value is an except because a card can't have several different mana 
# values. Winrate function finds suitable cards from dataframe and saves corresponding values to a list. 
# Then it calculates winrate.

point_rates = []
winrate = 0

def calculate_winrate(cards):
    for i in range(len(card_data)):
        for j in range(len(cards)):
            if card_data.iloc[i]['Card_name'] == cards[j]:
                point_rates.append(card_data.iloc[i]['Points_rate'])
    
    winrate = sum(point_rates)/len(point_rates)
    print(color, ',', mana, 'and', c_type, ' winrate: ', winrate)
    
    
if len(color_mana_type) != 0:
    calculate_winrate(color_mana_type)
    print('cmt', color_mana_type)
    print(len(color_mana_type))
    
elif len(color_mana) != 0:
    calculate_winrate(color_mana)
    print('cm', color_mana)
    print(len(color_mana))
    
elif len(color_type) != 0:
    calculate_winrate(color_type)
    print('ct', color_type)
    print(len(color_type))
    
elif len(mana_type) != 0:
    calculate_winrate(mana_type)
    print('mt', mana_type)
    print(len(mana_type))

elif (color_mana_type == []) & (color_mana == []) & (color_type == []) & (mana_type == []):
    # 1 input given.
    if (mana == 100) & (c_type == ['']):
        if len(color_list) > 1:
            calculate_winrate(color_list)
            print('c', color_list)
            print(len(color_list))
            
    if (color == ['']) & (mana == 100):
        if len(type_list) > 1:
            calculate_winrate(type_list)
            print('t', type_list)
            print(len(type_list))
    else:
        print('No cards with given features.')
```
